### Checking for outliers in the data per column:

```python
# Checking Outliers in dataset

col_names = list(df.columns)
fig, ax = plt.subplots(len(col_names), figsize=(8,100))

for i, col_val in enumerate(col_names):

    sns.boxplot(y=df[col_val], ax=ax[i])
    ax[i].set_title('Box plot - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)

plt.show()
```

As we can see above, almost all the variables have outliers present in them. Now let us go ahead and treat them.

```python
def treat_outlier(x):
    # taking 5,25,75 percentile of column
    q5= np.percentile(x,5)
    q25=np.percentile(x,25)
    q75=np.percentile(x,75)
    dt=np.percentile(x,95)
    #calculationg IQR range
    IQR=q75-q25
    #Calculating minimum threshold
    lower_bound=q25-(1.5*IQR)
    upper_bound=q75+(1.5*IQR)
    #Capping outliers
    return x.apply(lambda y: dt if y > upper_bound else y).apply(lambda y: q5 if y < lower_bound else y)
```

```python
for i in df:
    df[i]=treat_outlier(df[i])
```

### Calculating the Variance Inflation Factor to determine the variables used for prediction:

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

### Validating the model on the train and choosing the optimum cut-off:

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn import metrics
```

```python
y_predict_train = model_2.predict(X_train)
y_predict_train
```

```python
sns.boxplot(x = df['default'], y = y_predict_train)
plt.xlabel('default')
```

#### Cut-off is 0.05:

```python
y_class_pred1=[]
for i in range(0,len(y_predict_train)):
    if np.array(y_predict_train)[i]>0.05:
        a=1
    else:
        a=0
    y_class_pred1.append(a)
```

```python
sns.heatmap((metrics.confusion_matrix(y_train,y_class_pred1)),annot=True,fmt='.5g'
            ,cmap='plasma');
plt.xlabel('Predicted');
plt.ylabel('Actuals',rotation=0);
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_train,y_class_pred1).ravel()
print('True Negative:',tn,'\n''False Positives:' ,fp,'\n''False Negatives:', fn,'\n''True Positives:', tp)
```

```python
print(metrics.classification_report(y_train,y_class_pred1,digits=3))
```

#### Cut-off of 0.06:

```python
y_class_pred2=[]
for i in range(0,len(y_predict_train)):
    if np.array(y_predict_train)[i]>0.06:
        a=1
    else:
        a=0
    y_class_pred2.append(a)
```

```python
sns.heatmap(metrics.confusion_matrix(y_train, y_class_pred2), annot = True, fmt = '.5g', cmap = 'plasma');
plt.xlabel('Predicted')
plt.ylabel('Actuals', rotation = 0);
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_train,y_class_pred2).ravel()
print('True Negatives ', tn,'\n', 'False Positives ',fp,'\n','False Negatives ', fn,'\n','True Positives ',tp)
```

```python
print(metrics.classification_report(y_train,y_class_pred2,digits=3))
```

The recall values have actually decreased.

By changing the cut-off, we are able to significantly reduce the false negatives and increase the true positives, which in our business case, is the most important.
